# Loading the data

The `online_retail_II.xlsx` contains data regarding the sales of an online store.

```python
import pandas as pd
```

```python
df = pd.read_excel('online_retail_II.xlsx', sheet_name = None)
```

```python
df = pd.concat(df.values(), ignore_index = True)

df.head()
```

```python
df.info()
```

```python
# Set the new index as InvoiceDate

df.set_index('InvoiceDate', inplace = True)
df.head()
```

```python
df.describe()
```

```python
df[df['Quantity'] < 0]
```

```python
df[df['Price'] < 0]
```

```python
df.info()
```

```python
# Remove rows with negative quantities or price

df = df[(df['Quantity'] >= 0) & (df['Price'] >= 0)]

df.info()
```

```python
df.to_csv('sales.csv')
```
